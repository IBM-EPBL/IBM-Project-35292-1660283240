# Image Arithmetics



```
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
```

```
live = imread('/content/drive/MyDrive/IBM/Data Set/Dataset Plant Disease/fruit-dataset/fruit-dataset/test/Apple___Black_rot/04de1246-ecfa-4d2c-bc2d-16b81b0bd364___JR_FrgE.S 2970.JPG')
mask = imread('/content/drive/MyDrive/IBM/Data Set/Dataset Plant Disease/fruit-dataset/fruit-dataset/test/Apple___Black_rot/05fb01a0-fd64-4b3d-a512-c0bda181811e___JR_FrgE.S 2889.JPG')

plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(live, cmap='gray')
plt.subplot(122), plt.imshow(mask, cmap='gray')
plt.show()
```

```
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(live, cmap='gray')
plt.subplot(122), plt.imshow(live - 20, cmap='gray')
plt.show()
```

```
plt.figure(figsize=(10, 10))
plt.subplot(131), plt.imshow(mask - live, cmap='gray')
plt.subplot(132), plt.imshow(-(mask - live + 128), cmap='gray')
plt.subplot(133), plt.imshow(mask - live + 128, cmap='gray')
plt.show()
```

```
shaded = imread('/content/drive/MyDrive/IBM/Data Set/Dataset Plant Disease/fruit-dataset/fruit-dataset/test/Apple___Black_rot/05fb01a0-fd64-4b3d-a512-c0bda181811e___JR_FrgE.S 2889.JPG')
shading = imread('/content/drive/MyDrive/IBM/Data Set/Dataset Plant Disease/fruit-dataset/fruit-dataset/test/Apple___Black_rot/07dbfcca-71aa-4c14-b895-60dfb63f701d___JR_FrgE.S 3063.JPG')

plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(shaded, cmap='gray')
plt.subplot(122), plt.imshow(shading, cmap='gray')
plt.show()
```

```
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(np.multiply(shaded, 1/shading), cmap='gray')
plt.subplot(122), plt.imshow(shaded, cmap='gray')
plt.show()
```
